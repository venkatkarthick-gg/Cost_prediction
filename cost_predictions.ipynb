### Feature Engineering

```python
#converting age cat as under early till_25, till_35, above_35
```

```python
def age(x):

        if x<= 25:
            return('till_25')
        elif x>35:
            return('above 35')
        else:
            return('25-35')
```

```python
df['age_cat']=df['age'].apply(age)
```

```python
df['age_cat']
```

```python
#bmi categorizes into 3 subcategories
df['bmi']
```

```python
def bmi(x):

        if x<= 25:
            return('fit')
        elif x>35:
            return('Obese')
        else:
            return('fat')
```

```python
df['bmi_cat']=df['bmi'].apply(bmi)
```

```python
#lets try with other regression models
from sklearn.ensemble import RandomForestRegressor
from sklearn.tree import DecisionTreeRegressor
from sklearn.ensemble import AdaBoostRegressor
from xgboost import XGBRFRegressor
```

```python
#lets create a user defined function for fitting and evaluating the models.
performance=pd.DataFrame(columns=['model','r2_train','r2_test','mae_train','mae_test','mse_train','mse_test','remarks'])

def model(model,name,xtrain=xtrain,xtest=xtest):
    global performance
    model_r=model.fit(xtrain,ytrain)
    pred_train=model_r.predict(xtrain)
    pred_test=model_r.predict(xtest)
    
    mae_train=mean_absolute_error(ytrain,pred_train)
    mae_test=mean_absolute_error(ytest,pred_test)
    mse_train=mean_squared_error(ytrain,pred_train)
    mse_test=mean_squared_error(ytest,pred_test)
    r2_train=r2_score(ytrain,pred_train)
    r2_test=r2_score(ytest,pred_test)
    
    def remarks(r2_train,r2_test):
        if abs(r2_train-r2_test)>0.1 or r2_train >0.95:
            return('overfit')
        elif r2_train<0.77 or r2_test<0.77:
            return 'underfit'
        else:
            return 'goodfit'
    
    performance=performance.append({'model':name,'r2_train':r2_train,'r2_test':r2_test,
                                   'mae_train':mae_train,'mae_test':mae_test,'mse_train':mse_train,'mse_test':mse_test,
                                   'remarks':remarks(r2_train,r2_test)},
                                  ignore_index=True)
```

```python
model(RandomForestRegressor(),'Random_forest_regressor')
```

```python
performance
```

```python
# as we can see the model gets overfitted
```

```python
#Decision_tree_regressor
model(DecisionTreeRegressor(),'Decision_tree')
```

```python
performance
```

```python
#Adboost
model(AdaBoostRegressor(),'Ada_boost')
```

```python
performance
```

```python
#xgboost
model(XGBRFRegressor(),'Xgboost')
```

```python
performance
```
